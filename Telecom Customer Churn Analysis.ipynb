### Step 4: Replaceing Blank with Zero

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ",0)
df["TotalCharges"]=df["TotalCharges"].astype(float)
```

### Step 7: Convert 1 and 0 value of SeniorCitizen to Yes and NO

```python
def conv(value):
    if value==1:
        return "yes"
    else:
        return "no"
df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
```

```python
df.head()
```

### Step 8: Count of Customer by Churn 

```python
ax=sns.countplot(x=df["Churn"], data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customer by Churn")
plt.show()
```

### Step 9: Percentage of Churned Customers

```python
plt.figure(figsize=(3,4))
gb= df.groupby('Churn').agg({'Churn':'count'})
plt.pie(gb['Churn'], labels=gb.index, autopct="%1.2f%%")
plt.title("Percentage of Churned Customers",fontsize=10)
plt.show()
```

### Step 10: Churn by Gender

```python
plt.figure(figsize=(3,4))
sns.countplot(x="gender", data=df, hue='Churn')
plt.title('Churn by Gender')
plt.show()
```

### Step 11: Count of Customers by SeniorCitizen

```python
plt.figure(figsize=(3,4))
ax=sns.countplot(x="SeniorCitizen", data=df)
ax.bar_label(ax.containers[0])
plt.title('Count of Customers by SeniorCitizen')
plt.show()
```

```python
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

# Plot
fig, ax = plt.subplots(figsize=(6, 6))  # Adjust figsize for better visualization

# Plot the bars
total_counts.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e'])  # Customize colors

# Add percentage labels on the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center')

# Titles and labels
plt.title('Churn by Senior Citizen (Stacked Bar Chart)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Churn', loc='upper right')  # Customize legend location

# Show plot
plt.show()
```

### Step 14: Countplot for other values

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming your DataFrame is already loaded as df

# Column list
cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
        'PaperlessBilling', 'PaymentMethod', 'MonthlyCharges',
        'TotalCharges', 'Churn']

# Remove 'customerID' since it's unique
cols = [col for col in cols if col != 'customerID']

# Set up the subplot grid
n_cols = 4
n_rows = (len(cols) + n_cols - 1) // n_cols
plt.figure(figsize=(22, n_rows * 6))

for i, col in enumerate(cols):
    plt.subplot(n_rows, n_cols, i + 1)

    # Convert numerical variables to bins for countplot
    if df[col].dtype != 'object' and df[col].nunique() > 10:
        # Binning
        binned = pd.cut(df[col], bins=10)
        temp_df = df.copy()
        temp_df[col + '_binned'] = binned
        sns.countplot(data=temp_df, x=col + '_binned', hue='Churn')
        plt.xticks(rotation=45)
    else:
        sns.countplot(data=df, x=col, hue='Churn', order=df[col].value_counts().index)
        plt.xticks(rotation=45)

    plt.title(f'Countplot of {col} by Churn')

plt.tight_layout()
plt.show()
```

### The majority of customers who do not churn tend to have services like PhoneService, InternetService (particularly DSL), and OnlineSecurity enabled. For services like OnlineBackup, TechSupport, and StreamingTV, churn rates are noticeably higher when these services are not used or are unavailable.
